```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=500, noise=0.1)
```

#### K-Means Clustering & Results Plotted

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

km = KMeans(n_clusters=2, random_state=1).fit(X)

plt.scatter(X[:,0], X[:,1], c=km.labels_)
plt.show
```

#### DBSCAN clustering with eps=0.1 and min_samples=8 

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.1, min_samples=8).fit(X)

plt.scatter(X[:,0], X[:,1], c=db.labels_)
plt.show
```

#### DBSCAN clustering with eps=0.153 and min_samples=5 (Best Results)

```python
db = DBSCAN(eps=0.153, min_samples=5).fit(X)

plt.scatter(X[:,0], X[:,1], c=db.labels_)
plt.show
```

#### DBSCAN clustering with eps=0.163 and min_samples=8

```python
db = DBSCAN(eps=0.163, min_samples=8).fit(X)

plt.scatter(X[:,0], X[:,1], c=db.labels_)
plt.show
```

#### Results: K-Means Clustering fails to deal with non-cicular data. So the moon shaped data was not sperated into its culters well. However, on the other hand, DBSCAN clustering can handle different shaped data. Through my testing, an eps=0.153 and min_samples=5 gave me the best results.

#### Comparasion: DBSCAN requires two parameters while K-Means only needs 1. Furthermore, DBSCAN doesn't work well for sparse datasets, but can efficienly handle outliers and noisy datasets. While K-Means is not affected by spare datasets, but struggles with outliers and noisy data sets.
